## Redes acÃ­clicas

Um exemplo clÃ¡ssico de uma rede acÃ­clica dirigida Ã© uma rede de citaÃ§Ãµes de artigos, como a que foi referida na aula anterior. Ao escrever um artigo, o autor sÃ³ pode citar outro artigo se este jÃ¡ tiver sido escrito, o que significa que todas as arestas de uma rede de citaÃ§Ãµes apontam para trÃ¡s no tempo, de artigos posteriores para artigos anteriores.

NÃ£o Ã© Ã³bvio, mas nÃ£o deixa de ser verdade e pode ser demonstrado, que se uma rede for acÃ­clica poderÃ¡ ser desenhada da forma apresentada no slide anterior, com todas as arestas a apontar para baixo.

```python
G=nx.gnp_random_graph(10,0.5,directed=True)
DAG = nx.DiGraph([(u,v) for (u,v) in G.edges() if u < v])
nx.is_directed_acyclic_graph(DAG)
```

```python
nx.draw(DAG,pos = nx.circular_layout(DAG), arrowsize=20, node_color='white', edgecolors='black', node_size=500, 
        with_labels =True)
```

## ProjeÃ§Ã£o de redes Bipartidas

```python
from networkx.algorithms import bipartite
B = nx.Graph()
B.add_nodes_from([1, 2, 3, 4], bipartite=0)
B.add_nodes_from(["a", "b", "c"], bipartite=1)
B.add_edges_from([(1, "a"), (1, "b"), (2, "b"), (2, "c"), (3, "c"), (4, "a")])
nx.is_bipartite(B)
```

```python
bottom_nodes, top_nodes = bipartite.sets(B)
nx.draw(B, nx.bipartite_layout(B, top_nodes, align='horizontal'), arrowsize=20, node_color='white', edgecolors='black', node_size=500, 
        with_labels =True)
```

```python
from matplotlib import pyplot as plt
fig, axs = plt.subplots(ncols=2)
G1 = bipartite.projected_graph(B, bottom_nodes)
G2 = bipartite.projected_graph(B, top_nodes)
nx.draw(G1, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True, ax=axs[0])
nx.draw(G2, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True, ax=axs[1])
plt.show()
```

# Grau

O grau de um nÃ³ numa rede nÃ£o direcionada Ã© o nÃºmero de arestas ligadas a ele

$$k_i = \sum{n}_{j=1}A_{ij} $$

O nÃºmero total 2m das pontas das m arestas de uma rede Ã© dado por:

$$2m = \sum{n}_{i=1}k_{i}=\sum_{ij}A_{ij}$$

O grau mÃ©dio de uma rede Ã©:

$$c = \frac{1}{n}\sum{n}_{i=1}k_i=\frac{2m}{n}$$

## CÃ¡lculo do grau

## Caminhos mais curtos

```python
G = nx.Graph()
nx.add_path(G, [0, 1, 2])
nx.add_path(G, [0, 10, 2])
nx.add_path(G, [0, 5, 2])
nx.add_path(G, [5, 7, 2])
nx.draw(G, arrowsize=20, node_color='white', edgecolors='black', node_size=500, with_labels =True)
print([p for p in nx.all_shortest_paths(G, source=0, target=2)])
```
